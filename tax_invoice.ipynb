```python
import pdfplumber
import re
import pandas as pdllll


pdf_path = "/content/AUGUST INV DETAIL_compressed.pdf"

# Function to extract invoice details
def extract_invoice_details(pdf_path):
    invoice_data = []

    with pdfplumber.open(pdf_path) as pdf:
        invoice_no = None
        for page in pdf.pages:
            text = page.extract_text()

            if text:

                invoice_match = re.search(r"Invoice No\.\s*:\s*([\w\d]+)", text)
                if invoice_match:
                    invoice_no = invoice_match.group(1)


                lines = text.split("\n")
                for line in lines:
                    match = re.match(r"(\d+\.)\s+([\w\s\(\)-]+)\s+\d+\s+([\d\.]+)\s+C/S", line)
                    if match:
                        description = match.group(2).strip()
                        quantity = match.group(3).strip()
                        invoice_data.append([invoice_no, description, quantity])

    return invoice_data


data = extract_invoice_details(pdf_path)


df = pd.DataFrame(data, columns=["Invoice No", "Description of Goods", "Quantity"])


from IPython.display import display
display(df)
```

```python
import pdfplumber
import re
import pandas as pd

pdf_path = "/content/AUGUST INV DETAIL_compressed.pdf"

def extract_invoice_details_grouped(pdf_path):
    invoice_data = []
    grouped_data = []

    with pdfplumber.open(pdf_path) as pdf:
        invoice_no = None
        for page in pdf.pages:
            text = page.extract_text()

            if text:

                invoice_match = re.search(r"Invoice No\.\s*:\s*([\w\d]+)", text)
                if invoice_match:
                    if invoice_data:
                        grouped_data.extend(invoice_data)
                    invoice_no = invoice_match.group(1)
                    invoice_data = []


                lines = text.split("\n")
                for line in lines:
                    match = re.match(r"\s*\d+\.\s+(.+?)\s+\d+\s+([\d\.]+)\s+C/S", line)
                    if match:
                        description = match.group(1).strip()

                        description_cleaned = re.sub(r"\s*\d{6,}\s*", "", description)
                        quantity = match.group(2).strip()
                        invoice_data.append([invoice_no, description_cleaned, quantity])

        if invoice_data:
            grouped_data.extend(invoice_data)

    return grouped_data

grouped_data = extract_invoice_details_grouped(pdf_path)

df_grouped = pd.DataFrame(grouped_data, columns=["Invoice No", "Description of Goods", "Quantity"])

from IPython.display import display
display(df_grouped)
```

```python
import pandas as pd
excel_path = "Grouped_Invoice_Data.xlsx"
df_grouped.to_excel(excel_path, index=False)

print(f"Excel file saved successfully as {excel_path}")
```
